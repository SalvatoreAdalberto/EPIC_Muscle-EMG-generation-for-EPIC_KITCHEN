```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
from sklearn.manifold import TSNE
```

```python
num_clips = 5
```

```python
# EPIC-KITCHEN
# TODO: Add actionnet color and labels

labels = {'EK':{
        0 : "take (get)",
        1 : "put-down (put/place)",
        2 : "open",
        3 : "close",
        4 : "wash (clean)",
        5 : "cut",
        6 : "stir (mix)",
        7 : "pour"
}}

colors = {'EK': {
        0 : "#A52A2A",
        1 : "#DAA520",
        2 : "#FF7F50",
        3 : "#7BC8F6",
        4 : "#FFFF14",
        5 : "#76FF7B",
        6 : "#13EAC9",
        7 : "olive"
}}
```

```python
data = pd.DataFrame(pd.read_pickle('saved_features/reconstructed/VAE.pkl')['features_RGB'])

features = data['features']
features = [f[2] for f in features]
for i in range(len(features)):
    if len(features[i]) != 1024:
        print(f"OPS, PROBLEMA: {len(features[i])}")

reduced = TSNE().fit_transform(features)
data['x'] = reduced[:, 0]
data['y'] = reduced[:, 1]
for i in range(8): # ek has 8 classes
    filtered = data[data["label"] == i]
    plt.scatter(filtered['x'], filtered['y'], c=colors['EK'][i], label=labels['EK'][i])
# plt.legend()
plt.show()
```

```python
def visualize_features(title, filename, annotation, video_level=True, dataset='EK', save=True):
    data = pd.DataFrame(pd.read_pickle(filename)['features'])
    annotations = pd.read_pickle(annotation)
    data = pd.merge(data, annotations, how="inner", on="uid")
    features = data['features_RGB']
    if video_level:
        features = [f[num_clips//2] for f in features]
    reduced = TSNE().fit_transform(features)
    data['x'] = reduced[:, 0]
    data['y'] = reduced[:, 1]
    for i in range(8): # ek has 8 classes
        filtered = data[data["verb_class"] == i]
        plt.scatter(filtered['x'], filtered['y'], c=colors[dataset][i], label=labels[dataset][i])
    plt.legend()
    plt.title(title)
    plt.savefig(f"img/{filename.split('/')[1]}.png")
    plt.show()
```

```python
visualize_features("Init on kinetics dataset", "saved_features/PT_D_16f_5c_D1_train.pkl", "train_val/D1_train.pkl")
```

```python
visualize_features("Finetuned Epic Kitchen", "saved_features/FT_D_16f_5c_D1_train.pkl", "train_val/D1_train.pkl")
```
