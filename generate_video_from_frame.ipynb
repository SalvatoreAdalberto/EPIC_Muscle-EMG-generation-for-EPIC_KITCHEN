```python
ANNOTATION_DIR = 'train_val/'
FRAMES_DIR = '../action_net/RGB/ActionNet_S04/'
import pandas as pd
import os

df = pd.read_pickle(ANNOTATION_DIR + 'S04_test.pkl')
for i in range(len(df)):
    sample = df.iloc[i]
    os.system(f"ffmpeg -framerate 30 -start_number {sample.start_frame} -i {FRAMES_DIR}img_%10d.jpg -vframes {sample.stop_frame - sample.start_frame} -vcodec mpeg4 ../test_video/{sample.verb + '_' + str(sample.uid)}.avi")
```
